```python
import pandas as pd
from pandas import Series,DataFrame

import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns


import sklearn
from sklearn.linear_model import LinearRegression
```

```python
titanic_df = pd.read_csv('train.csv')
```

```python
def male_female_child(passenger):
    # 年齢と性別のデータを取得します。
    age,sex = passenger
    # 年齢を調べて16歳未満なら、子供。それ以外は性別をそのまま返します。
    if age < 16:
        return 3
    else:
        if sex == 'male':
            return 1
        else:
            return 2
    
# personという新しい列を追加します。
titanic_df['Person'] = titanic_df[['Age','Sex']].apply(male_female_child,axis=1)
```

```python
def is_cabin(cabin):
    #print(int(cabin.notnull()))
    return int(cabin.notnull())
    
titanic_df['Is_cabin'] = titanic_df[['Cabin']].apply(is_cabin,axis=1)
```

```python
titanic_df['Alone'] = titanic_df.Parch + titanic_df.SibSp
```

```python
titanic_df['Alone'].loc[titanic_df['Alone'] >0] = 1
```

```python
#drop column
train_data = titanic_df.drop(['Name', 'Ticket', 'Cabin', 'Embarked'], 1)
```

```python
def get_trani_data_on_age(train_data, sex):
    train_data_age = train_data[train_data.Sex == sex]
    train_data_age = train_data_age.drop('Sex', 1)
    
    # train male Age
    X_data = train_data_age[train_data_age.Age.notnull()].drop(['Survived', 'Age'], 1)
    Y_age = train_data_age[train_data_age.Age.notnull()].Age 
    
    age_lreg = LinearRegression()
    age_lreg.fit(X_data,Y_age)
    
    X_predict = train_data_age[train_data_age.Age.isnull()].drop(['Survived', 'Age'], 1)
    Y_target = age_lreg.predict(X_predict)
    Y_target = np.round(Y_target)
    
    X_predict['Survived'] = train_data_age[train_data_age.Age.isnull()].Survived
    X_predict['Age'] = Y_target
    
    train_data_on_age = pd.concat([train_data_age[train_data_age.Age.notnull()], X_predict])
    
    return age_lreg, train_data_on_age
```

```python
#ses age
age_lreg_male, train_data_male = get_trani_data_on_age(train_data, 'male')
age_lreg_female, train_data_female = get_trani_data_on_age(train_data, 'female')
```

```python
train_data_age = train_data[train_data.Sex == 'male']
train_data_age = train_data_age.drop('Sex', 1)
    
# train male Age
X_data = train_data_age[train_data_age.Age.notnull()].drop(['Survived', 'Age'], 1)
Y_age = train_data_age[train_data_age.Age.notnull()].Age

pred_age = age_lreg_male.predict(X_data)
```

```python
def get_survived_model(train_data):
    X_data = train_data.drop('Survived', 1)
    Y_survived = train_data.Survived
    
    survived_lreg = LinearRegression()
    survived_lreg.fit(X_data,Y_survived)
    
    return survived_lreg
```

```python
survived_lreg_male = get_survived_model(train_data_male)
survived_lreg_female = get_survived_model(train_data_female)
```
